```python
for col in ['discounted_price',	'actual_price',	'discount_percentage','rating',	'rating_count']:
  df[col]=df[col].astype(str).str.replace('₹','').str.replace('%','').str.replace(',','').str.replace('nan','0')
  df[col] = pd.to_numeric(df[col].str.replace('[^0-9.]','',regex=True),errors='coerce').fillna(0)
```

```python
df.isnull().sum()
```

```python
df.describe(include="O")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='rating', palette='deep')
plt.title('Number of rating')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df, x="discount_percentage", color="chocolate", kde=True)
plt.xlabel("discount_percentage")
plt.ylabel("Frequancy")
plt.title("Discount Percentage Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df, x="discounted_price", color="gold", kde=True)
plt.xlabel("discount_price")
plt.ylabel("Frequancy")
plt.title("Discount Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df, x="rating", color="darkred", kde=True)
plt.xlabel("rating")
plt.ylabel("Frequancy")
plt.title("rating Distribution")
plt.show()
```

```python
discount_percentage_counts = df['discount_percentage'].value_counts().head(10)
plt.pie(discount_percentage_counts.values, labels=discount_percentage_counts.index, autopct='%1.2f%%')
plt.title('Top 10 discount_percentage Distribution')
plt.legend(bbox_to_anchor=(1.1, 1))
plt.show()
```

```python
plt.figure(figsize=(8,6))
numerical_cols=['rating','rating_count','actual_price','discount_percentage','discounted_price']
corr_matrix=df[numerical_cols].corr()
sns.heatmap(corr_matrix,annot=True,fmt='.2f',cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='discount_percentage',y='discounted_price',data=df,marker="o", color="darkblue")
plt.xlabel('discount_percentage')
plt.ylabel('discounted_price')
plt.title("Discounted Price vs. Discount Percentage")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='rating',y='discounted_price',data=df,marker="o", color="dimgrey")
plt.xlabel('rating')
plt.ylabel('discounted_price')
plt.title("Discounted Price vs. Rating")
plt.show()
```
